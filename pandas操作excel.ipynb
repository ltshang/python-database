```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
df = DataFrame({'类别':['水果','水果','水果','蔬菜','蔬菜','肉类','肉类'],
                '产地':['美国','中国','中国','中国','新西兰','新西兰','美国'],
                '水果':['苹果','梨','草莓','番茄','黄瓜','羊肉','牛肉'],
               '数量':[5,5,9,3,2,10,8],
               '价格':[5,5,10,3,3,13,20]})
print(df)
```

```python
print(pd.crosstab(df['类别'],df['产地'],margins=False))
```

```python
print(pd.crosstab(df['类别'],df['产地'],margins=True))
```

```python
df['价格'] = pd.cut(df['价格'],[0,10,20,30],labels = ['小于10','10-20','20-30'])
df['价格']
```

```python
df['数量'] = pd.cut(df['数量'],[0,5,10],labels = ['小于5','5-10'])
df['数量']
```

```python
s = pd.concat([df['价格'] ,df['数量']],axis = 1)
s
```

```python
print(pd.crosstab(df['价格'],df['数量'],margins=True))
```

```python
d = pd.crosstab(df['价格'],df['数量'],margins = False)
d
```
